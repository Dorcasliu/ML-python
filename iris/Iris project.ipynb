## supervised learning: Simple linear regression 

```python
x=10*np.random.rand(100)
```

```python
y=-5*x+np.random.randn(100)
```

```python
plt.scatter(x,y)
```

```python
import sklearn
print(sklearn.__version__)
from sklearn.linear_model import LinearRegression
```

```python
# CHOOSE PARAMATERS
model=LinearRegression(fit_intercept=True,copy_X=True)
model
```

```python
X=x.reshape(-1,1) #feature matrix:X
X.shape
```

```python
#4.
model.fit(X,y)
#model.coef_
```

```python
model.coef_
```

```python
model.intercept_
```

```python
x_fit=np.linspace(-1,11) #range from -1 till 11, default num:50.
```

```python
X_fit=x_fit.reshape(-1,1) #1 column
```

```python
y_fit=model.predict(X_fit)
```

```python
plt.scatter(x,y)
plt.plot(x_fit, y_fit)
```
